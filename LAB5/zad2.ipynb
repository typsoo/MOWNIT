```python
import numpy as np
import matplotlib.pyplot as plt
import PIL


image = PIL.Image.open('Lenna_gray.jpg') 


U, S, Vt = np.linalg.svd(np.array(image), full_matrices=False)
k_values = [1, 2, 5, 10, 30, 512]

fig, axes = plt.subplots(1, len(k_values) + 1, figsize=(18, 5))

axes[0].imshow(image, cmap='gray')
axes[0].set_title(f'Original\nRank: {512}')
axes[0].axis('off')

for i, k in enumerate(k_values):
    compressed_img = U[:, :k] @ np.diag(S[:k]) @ Vt[:k, :]
    
    
    axes[i+1].imshow(compressed_img, cmap='gray')
    axes[i+1].set_title(f'Compressed\nk = {k}')
    axes[i+1].axis('off')

plt.tight_layout()
plt.show()


```
